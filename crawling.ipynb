```python
from multiprocessing import Pool 
from selenium import webdriver
import urllib.request
import time
import os 

pageNum=2
imgNum=0
img_folder = './img'
img_url = []
if not os.path.isdir(img_folder) : 
    os.mkdir(img_folder)

driver = webdriver.Chrome("./chromedriver.exe") #자기 크롬드라이버 있는 경로

def recursive(imgsrc):
    global imgNum
    imgsrc.click()
    time.sleep(1.5)
    imgUrl = driver.find_elements_by_css_selector("body > div > div.l-body > div.l-layout.l-layout_tight > div > div.content-main > div.wallpaper > div.wallpaper__placeholder > img")
    try:
        src = imgUrl[0].get_attribute('src')
        print(src)
        urllib.request.urlretrieve(src, f'{img_folder}/{imgNum}.jpg')
        imgNum+=1
        driver.back()
    except AttributeError:
        import pdb;pdb.set_trace()
    
while pageNum < 126:
    url = "https://wallpaperscraft.com/catalog/space/page" #자기가 크롤링할 페이지
    url = url + str(pageNum)
    driver.get(url)
 
    images = driver.find_elements_by_css_selector("img.wallpapers__image")

    for img in images:
        time.sleep(1)
        recursive(img)
    pageNum += 1
```
